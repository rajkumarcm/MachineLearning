```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
N = 100
x1 = np.linspace(-4*np.pi, 4*np.pi, 100)
x2 = np.sin(x1)
X = np.c_[x1, x2]
```

```python
U, sigma, VT = np.linalg.svd(X)
S = np.zeros([100, 2])
S[:2, :2] = np.diag(sigma)
Z = U @ S
X_recon = Z @ VT
```

```python
plt.figure()
plt.plot(x1, x2, '.b')
plt.figure()
plt.plot(X_recon[:, 0], X_recon[:, 1], '.r')
```
