```python
import numpy as np
from scipy.stats import beta, binom
%matplotlib inline
from matplotlib import pyplot as plt
np.random.seed(123)
```

```python
a = 9
b = 3
theta = np.random.beta(a=a+2, b=b+8, size=[100])
theta = np.sort(theta)
density = beta.pdf(theta, a=a+2, b=b+8)
plt.figure()
plt.title("Posterior distribution after having seen 2 heads, and 8 tails")
plt.xlabel("$\Theta$")
plt.ylabel("$p(\Theta|D)$")
plt.plot(theta, density, '-m')
plt.show()
```

```python
x = np.zeros([1000])
for i in range(1000):
    ran_theta = np.random.beta(a=a+2, b=b+8, size=[1])
    x[i] = np.random.binomial(n=10, p=ran_theta, size=[1])
plt.figure()
plt.title("Posterior predictive distribution approximated by sampling")
plt.xlabel("Heads sampled for future trials")
plt.ylabel("Frequency of heads based on sampled $p$")
plt.hist(x, rwidth=0.8)
plt.show()
```

```python
# For personal reference

prior_x = np.random.beta(a=201, b=801, size=100)
prior_x = np.sort(prior_x)
prior_y = beta.pdf(prior_x, a=201, b=801)
plt.figure()
plt.title("Plot of prior distribution beta")
plt.xlabel("$\Theta$ ~ Beta(a=201, b=801)")
plt.ylabel("$Beta(\Theta|a=201, b=801)$")
plt.plot(prior_x, prior_y, '-b')
plt.show()
```

```python
# Posterior predictive distribution by plugging in Posterior Mean

theta = (a + 2)/(a + b + 10)
x = np.random.binomial(n=10, p=theta, size=[1000])
plt.figure()
plt.title("Posterior predictive distribution approximated by sampling")
plt.xlabel("Heads sampled for future trials")
plt.ylabel("Frequency of heads based on sampled $p$")
plt.hist(x, rwidth=0.8)
plt.show()
```
